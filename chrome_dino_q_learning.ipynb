```python
class Agent(nn.Module):
    def __init__(self,n_actions,epsilon = 0):
        
        super().__init__()
        self.n_actions = n_actions
        self.epsilon = epsilon
        
        self.conv1 = nn.Conv2d(4,16,3, stride = 2)
        self.conv2 = nn.Conv2d(16,32,3, stride =2)
        self.conv3 = nn.Conv2d(32,64,3, stride = 2)
        self.linear1 = nn.Linear(94464,256)
        self.linear2 = nn.Linear(256,128)
        self.linear3 = nn.Linear(128,n_actions)
        
        
    
    def forward(self, state):
        x = F.relu(self.conv1(state))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = x.view(x.shape[0],-1)
#         print(x.shape)
        
        x = self.linear1(x)
        x = self.linear2(x)
        qvalues = self.linear3(x)
        
        return qvalues
    
    def get_qvalues(self,states):
        
#         model_device = next(self.parameters()).device
#         states = torch.tensor(states, device=model_device, dtype=torch.float)
        qvalues = self.forward(torch.tensor(states))
        print(qvalues)
        return qvalues.data.cpu().numpy()
    
    def get_action(self, qvalues):
        epsilon = self.epsilon
        
        random_action = np.random.choice(range(n_actions))
        best_action = qvalues.argmax(axis = -1)
        
        explore_or_not = np.random.choice([0,1],p=[1-epsilon,epsilon])
        
        return np.where(explore_or_not, random_action, best_action)
        
```

```python
def play(n_steps,agent,exp_replay):
#     start = time.time()
    is_done,state = getFrames() #takes around 1 second
    rewards = 0
    for _ in range(n_steps):
        print('\n\nStep: ',_)
        if is_done:
            print('Game Over')
            pyautogui.press('space')
            time.sleep(1)
            is_done,state = getFrames()
            continue
#         start = time.time()
        q = agent.get_qvalues([state]) #takes 0.14 seconds to predict
#         print(time.time()-start)
#         start = time.time()
        a = agent.get_action(q) #takes 0.0009 seconds to choose an action, takes around 1.4 seconds to get ss and take action
        print('Chosen action: ',a[0])
        if a[0]==0:
            print('space\n')
            pyautogui.press('space')
        else:
            print('nothing\n')
#         print('Time from ss to taking an action',time.time() - start)
        is_done,next_states = getFrames() #takes oround one second
        if is_done:
            reward = -100
        else:
            reward = 1
        exp_replay.add(state,a[0],reward,next_states,is_done)
        rewards+=reward
        state = next_states
#         elif a==1:
#             pyautogui.press('down')
            
#     print(time.time() - start)
    return rewards     
```

```python
def compute_loss(states,actions,rewards,next_states,is_done,agent,target_network,gamma = 0.99, device = device):
    states = torch.tensor(states, device=device, dtype=torch.float)
    
    actions = torch.tensor(actions, device=device, dtype=torch.long)
    rewards = torch.tensor(rewards, device=device, dtype=torch.float)
    
    next_states = torch.tensor(next_states, device=device, dtype=torch.float)
    
    is_done = torch.tensor(is_done.astype('float32'), device=device, dtype=torch.float)
    is_not_done = 1 - is_done
    
    predicted_qvalues = agent(states)
#     print('Agent:',predicted_qvalues)
    predicted_next_qvalues = target_network(next_states)
#     print('Agent:',predicted_next_qvalues)
    
    predicted_qvalues_for_actions = predicted_qvalues[range(
        len(actions)), actions]
    
    gamma_times_next_values = torch.tensor([gamma * max(p) for p in predicted_next_qvalues],device = device)
    
    reward_plus_gamma_times = rewards + is_not_done*gamma_times_next_values
    
    loss = torch.mean((predicted_qvalues_for_actions - reward_plus_gamma_times.detach())**2)
    
    
    return loss
```

```python
def evaluate(n_games,agent,time_steps):
    rw = []
    for _ in range(n_games):
        pyautogui.press('space')
        time.sleep(0.1)
        state = getFrames()
        reward = 0
        for _ in range(time_steps):
            if state == None:
                break
            q = agent.get_qvalues([state])
            a = agent.get_action(q)
            if a[0]==0:
                print('space\n')
                pyautogui.press('space')
            else:
                print('nothing\n')
            next_states = getFrames()
            if next_states != None:
                exp_replay.add(state,a[0],1,next_states,False)
            state = next_states
#         elif a==1:
#             pyautogui.press('down')

            reward+=1
        rw.append(reward)
    return np.mean(rw)
    
```
